* Gerar dados aleatórios e plotar um histograma (distribuição)


```python
minutos_dia = np.random.randint(0, 60, 30) # Gerar 30 valores, sendo 1 por dia, com até 60 min.
minutos_dia
```

```python
ax = sns.histplot(minutos_dia,) # Gera um histograma, que é um gráfico de distribuição.
ax.set(xlabel = 'Minutos por dia', ylabel = 'Qtde Vezes')
plt.show()
```

* Gerar uma matriz de correlação (de Person) de variáveis
Lembrando que valor 1 é a correlação direta perfeita e -1 é correlação negativa perfeita e valores próximos a zero indicam a falta de existência ou dificuldade de estabelecer correlação.

```python
mask = np.triu(np.ones_like(dados.corr(), dtype=bool))
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(dados.corr(), mask = mask, square = True, annot=True, vmin=-1, vmax=1 )
plt = show()
```
